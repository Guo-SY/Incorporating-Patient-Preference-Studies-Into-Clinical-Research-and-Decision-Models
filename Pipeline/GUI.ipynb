```python
import ipywidgets as widgets
from IPython.display import display

# Funzione per gestire l'azione del pulsante
def on_button_click(b):
    # Recupera i valori dai widget
    input_path_value = input_path.value
    output_path_value = output_path.value
    selected_format_value = format_checkbox.value

    # Fai qualcosa con i valori (in questo caso, li stampo)
    print(f"Input Path: {input_path_value}")
    print(f"Output Path: {output_path_value}")
    print(f"Selected Format: {selected_format_value}")

    diz_chart=main_chart(input_path=input_path_value, output_path=output_path_value)
    diz_tab=main_tables(input_path=input_path_value, output_path=output_path_value, mode="analysis")
    print(diz_chart)
    print(diz_tab)
    for i in diz_tab.keys():
      for j in diz_tab[i]:
        display(j)

# Widget per il percorso di input
input_path = widgets.Text(description="Input Path:")

# Widget per il percorso di output
output_path = widgets.Text(description="Output Path:")

# Widget per scegliere il formato
format_checkbox = widgets.Checkbox(value=False, description="Format: CSV")
format_checkbox_e = widgets.Checkbox(value=False, description="Format: Excel")
# Pulsante per eseguire l'azione
button = widgets.Button(description="Esegui")
button.on_click(on_button_click)

# Organizza i widget in un layout
layout = widgets.VBox([input_path, output_path, format_checkbox,format_checkbox_e, button])

# Mostra il layout
display(layout)
```
